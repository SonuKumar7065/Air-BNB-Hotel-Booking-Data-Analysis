```python
# ============================================================
# 🏨 AIRBNB HOTEL BOOKING DATA ANALYSIS
# Author: Sonu Kumar
# Internship: VOIS & Vodafone Idea Foundation - Conversational Data Analytics
# ============================================================

# 0. Setup
import os, warnings
warnings.filterwarnings("ignore")
import pandas as pd, numpy as np
import matplotlib.pyplot as plt, seaborn as sns
sns.set_theme(style="whitegrid", palette="Set2")

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import (
    mean_absolute_error, mean_squared_error, r2_score,
    accuracy_score, precision_score, recall_score, f1_score,
    confusion_matrix, roc_auc_score, classification_report
)

RSEED = 42
os.makedirs("outputs", exist_ok=True)
os.makedirs("dataset", exist_ok=True)

# 1. Load dataset (try common paths -> prompt upload if missing)
possible_names = ["Airbnb_Open_Data.xlsx", "/content/Airbnb_Open_Data.xlsx", "/mnt/data/1730285881-Airbnb_Open_Data.xlsx"]
file_path = None
for p in possible_names:
    if os.path.exists(p):
        file_path = p
        break

if file_path is None:
    # Ask for upload (Colab)
    try:
        from google.colab import files
        print("Please upload your 'Airbnb_Open_Data.xlsx' file when prompted...")
        uploaded = files.upload()
        file_path = list(uploaded.keys())[0]
    except Exception:
        raise FileNotFoundError("Dataset file not found. Upload 'Airbnb_Open_Data.xlsx' to the working directory.")

print("Loading dataset from:", file_path)
df = pd.read_excel(file_path)
print("Initial shape:", df.shape)
display(df.head())

# 2. Normalize column names
df.columns = [c.strip().lower().replace(" ", "_") for c in df.columns]
print("Columns:", df.columns.tolist())

# 3. Quick info & missing values
print("\n--- Data Info ---")
display(df.info())
print("\n--- Missing values (top 20) ---")
display(df.isnull().sum().sort_values(ascending=False).head(20))

# 4. Basic cleaning / numeric conversion
def safe_numeric(colname):
    if colname in df.columns:
        df[colname] = df[colname].astype(str).str.replace(r'[^0-9\.\-]', '', regex=True)
        df[colname] = pd.to_numeric(df[colname], errors='coerce')

for col in ['price','service_fee','minimum_nights','number_of_reviews','reviews_per_month',
            'review_rate_number','calculated_host_listings_count','availability_365','construction_year']:
    safe_numeric(col)

if 'last_review' in df.columns:
    df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')

# Fill missing values: numeric -> median, categorical -> 'unknown'
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
cat_cols = df.select_dtypes(include=['object']).columns.tolist()
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
df[cat_cols] = df[cat_cols].fillna("unknown")

# Save cleaned dataset
cleaned_path = "dataset/cleaned_airbnb.csv"
df.to_csv(cleaned_path, index=False)
print("Saved cleaned dataset:", cleaned_path)

# 5. Answer the QUESTIONS one-by-one with code, tables & plots
# Create helper save_plot
def save_fig(fig, fname):
    path = os.path.join("outputs", fname)
    fig.savefig(path, bbox_inches="tight", dpi=200)
    print("Saved:", path)

print("\n\n==================== QUESTIONS ANSWERS ====================\n")

# Q1: What are the different property types in the Dataset?
print("Q1: Property types (room_type counts):")
if 'room_type' in df.columns:
    rt_counts = df['room_type'].value_counts()
    display(rt_counts.to_frame("count"))
    fig, ax = plt.subplots(figsize=(6,4))
    sns.barplot(x=rt_counts.values, y=rt_counts.index, palette="viridis", ax=ax)
    ax.set_title("Room Type Counts")
    save_fig(fig, "q1_room_type_counts.png")
    plt.show()
else:
    print("Column 'room_type' not found.")

# Q2: Which neighborhood group has the highest number of listings?
print("\nQ2: Neighborhood group with highest number of listings:")
if 'neighbourhood_group' in df.columns:
    ng_counts = df['neighbourhood_group'].value_counts()
    display(ng_counts.to_frame("listings_count"))
    fig, ax = plt.subplots(figsize=(8,4))
    sns.barplot(x=ng_counts.index, y=ng_counts.values, palette="magma", ax=ax)
    ax.set_title("Listings by Neighbourhood Group")
    ax.set_ylabel("Number of listings")
    save_fig(fig, "q2_neighbourhood_group_listings.png")
    plt.show()
    top_ng = ng_counts.idxmax()
    print("Neighborhood group with highest listings:", top_ng)
else:
    print("Column 'neighbourhood_group' not found.")

# Q3: Which neighbourhoods have the highest average prices for Airbnb listings?
print("\nQ3: Top neighbourhoods by average price:")
if 'neighbourhood' in df.columns:
    nb_avg = df.groupby('neighbourhood')['price'].agg(['mean','median','count']).sort_values('mean', ascending=False)
    display(nb_avg.head(15))
    fig, ax = plt.subplots(figsize=(8,6))
    top10 = nb_avg[nb_avg['count']>=5].head(10)  # require at least 5 listings
    sns.barplot(x=top10['mean'], y=top10.index, palette="cividis", ax=ax)
    ax.set_xlabel("Mean Price")
    ax.set_title("Top 10 Neighbourhoods by Mean Price (min 5 listings)")
    save_fig(fig, "q3_top10_neighbourhoods_mean_price.png")
    plt.show()
else:
    print("Column 'neighbourhood' not found.")

# Q4: Is there a relationship between the construction year of property and price?
print("\nQ4: Relationship between construction_year and price:")
if 'construction_year' in df.columns:
    # group by decade and compute median price
    df['construction_year'] = pd.to_numeric(df['construction_year'], errors='coerce')
    df = df[df['construction_year'].notnull()]  # drop if none
    df['construction_decade'] = (df['construction_year']//10)*10
    cons_stats = df.groupby('construction_decade')['price'].agg(['median','count']).sort_index()
    display(cons_stats.head(15))
    fig, ax = plt.subplots(figsize=(8,4))
    cons_stats['median'].plot(marker='o', ax=ax)
    ax.set_title("Median Price by Construction Decade")
    ax.set_xlabel("Construction Decade")
    ax.set_ylabel("Median Price")
    save_fig(fig, "q4_construction_vs_price.png")
    plt.show()
else:
    print("Column 'construction_year' not found in dataset.")

# Q5: Who are the top 10 hosts by calculated host listing count?
print("\nQ5: Top 10 hosts by calculated_host_listings_count:")
if 'host_id' in df.columns and 'calculated_host_listings_count' in df.columns:
    host_top = df.groupby(['host_id','host_name'])['calculated_host_listings_count'].max().sort_values(ascending=False).head(10)
    display(host_top.to_frame("listings_count"))
    fig, ax = plt.subplots(figsize=(8,4))
    host_top.plot(kind='bar', ax=ax)
    ax.set_ylabel("Listings Count")
    ax.set_title("Top 10 Hosts by Listings Count")
    save_fig(fig, "q5_top10_hosts_listings.png")
    plt.show()
else:
    print("Required host columns not found ('host_id'/'calculated_host_listings_count').")

# Q6: Are hosts with verified identities more likely to receive positive reviews?
print("\nQ6: Host identity verification vs median review_rate_number:")
if 'host_identity_verified' in df.columns and 'review_rate_number' in df.columns:
    ver_med = df.groupby('host_identity_verified')['review_rate_number'].median().sort_values(ascending=False)
    display(ver_med.to_frame("median_review_rate"))
    fig, ax = plt.subplots(figsize=(6,3))
    sns.barplot(x=ver_med.index, y=ver_med.values, ax=ax)
    ax.set_title("Median Review Rate by Host Identity Verified")
    save_fig(fig, "q6_host_verified_vs_reviews.png")
    plt.show()
else:
    print("Columns 'host_identity_verified' or 'review_rate_number' missing.")

# Q7: Is there a correlation between the price of a listing and its service fee?
print("\nQ7: Correlation price vs service_fee:")
if 'service_fee' in df.columns and 'price' in df.columns:
    corr = df[['price','service_fee']].corr().iloc[0,1]
    print("Pearson correlation (price, service_fee):", corr)
    fig, ax = plt.subplots(figsize=(6,4))
    sns.scatterplot(x='price', y='service_fee', data=df.sample(min(len(df),2000), random_state=RSEED), alpha=0.5)
    ax.set_xlim(0, df['price'].quantile(0.99))
    save_fig(fig, "q7_price_vs_service_fee.png")
    plt.show()
else:
    print("Columns 'price' or 'service_fee' missing.")

# Q8: What is the average review rate number (e.g., stars), and does it vary by neighbourhood_group and room_type?
print("\nQ8: Average review rate & variation by neighbourhood_group and room_type:")
if 'review_rate_number' in df.columns:
    overall_mean = df['review_rate_number'].mean()
    print("Overall average review_rate_number:", round(overall_mean,3))
    # by neighbourhood_group
    if 'neighbourhood_group' in df.columns:
        grp = df.groupby('neighbourhood_group')['review_rate_number'].agg(['mean','count']).sort_values('mean',ascending=False)
        display(grp)
        fig, ax = plt.subplots(figsize=(8,4))
        sns.barplot(x=grp.index, y=grp['mean'], palette="rocket", ax=ax)
        ax.set_title("Average Review Rate by Neighbourhood Group")
        save_fig(fig, "q8_review_by_neighbourhood_group.png")
        plt.show()
    # by room_type
    if 'room_type' in df.columns:
        rt = df.groupby('room_type')['review_rate_number'].agg(['mean','count']).sort_values('mean',ascending=False)
        display(rt)
        fig, ax = plt.subplots(figsize=(6,4))
        sns.barplot(x=rt.index, y=rt['mean'], palette="viridis", ax=ax)
        ax.set_title("Average Review Rate by Room Type")
        save_fig(fig, "q8_review_by_room_type.png")
        plt.show()
else:
    print("Column 'review_rate_number' missing.")

# Q9: Are hosts with a higher calculated host listings count more likely to maintain higher availability throughout the year?
print("\nQ9: Listings count vs availability (365):")
if 'calculated_host_listings_count' in df.columns and 'availability_365' in df.columns:
    # compute median availability by host listing count bucket
    df['host_listing_bucket'] = pd.cut(df['calculated_host_listings_count'], bins=[-1,1,3,10,100], labels=["1","2-3","4-10","10+"])
    avail = df.groupby('host_listing_bucket')['availability_365'].median()
    display(avail.to_frame("median_availability"))
    fig, ax = plt.subplots(figsize=(6,4))
    avail.plot(kind='bar', ax=ax)
    ax.set_title("Median Availability (days/year) by Host Listing Bucket")
    save_fig(fig, "q9_availability_by_host_listings.png")
    plt.show()

    # scatter: listings count vs availability (median per host)
    host_avail = df.groupby('host_id').agg(listings=('calculated_host_listings_count','max'),
                                            median_avail=('availability_365','median')).reset_index()
    host_avail = host_avail[host_avail['listings'].notnull()]
    host_avail_sample = host_avail.sample(min(len(host_avail),2000), random_state=RSEED)
    fig, ax = plt.subplots(figsize=(6,4))
    sns.scatterplot(x='listings', y='median_avail', data=host_avail_sample, alpha=0.6)
    ax.set_xlabel("Listings (per host)")
    ax.set_ylabel("Median availability (days/year)")
    ax.set_title("Host listings vs median availability")
    save_fig(fig, "q9_host_listings_vs_availability.png")
    plt.show()
else:
    print("Required columns for Q9 missing.")

# ---------------------------
# Optional: Quick modeling (regression + classification) but optimized
# ---------------------------
print("\n--- OPTIONAL: Quick Models (optimized) ---")
# Build X, features as before (must exist)
model_features = []
for f in ['room_type','neighbourhood_group','neighbourhood','instant_bookable','cancellation_policy',
          'minimum_nights','number_of_reviews','reviews_per_month','review_rate_number',
          'calculated_host_listings_count','availability_365','host_identity_verified']:
    if f in df.columns:
        model_features.append(f)

if len(model_features) >= 2 and 'price' in df.columns:
    X = df[model_features].copy()
    y_reg = df['price'].copy()
    y_clf = (df['price'] > df['price'].median()).astype(int)

    # reduce neighbourhood cardinality
    if 'neighbourhood' in X.columns:
        top20 = X['neighbourhood'].value_counts().nlargest(20).index
        X['neighbourhood'] = X['neighbourhood'].apply(lambda v: v if v in top20 else 'Other')

    num_feats = [c for c in X.columns if X[c].dtype in [np.float64, np.int64]]
    cat_feats = [c for c in X.columns if c not in num_feats]

    try:
        encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
    except TypeError:
        encoder = OneHotEncoder(handle_unknown='ignore', sparse=False)

    preproc = ColumnTransformer(transformers=[
        ('num', StandardScaler(), num_feats),
        ('cat', encoder, cat_feats)
    ], remainder='drop')

    # Regression quick
    reg_pipe = Pipeline([('pre', preproc), ('rf', RandomForestRegressor(n_estimators=80, max_depth=10, n_jobs=-1, random_state=RSEED))])
    Xtr, Xte, ytr, yte = train_test_split(X, y_reg, test_size=0.2, random_state=RSEED)
    print("Training regression model (fast RF) ...")
    reg_pipe.fit(Xtr, ytr)
    y_pred = reg_pipe.predict(Xte)
    mae = mean_absolute_error(yte, y_pred)
    rmse = np.sqrt(mean_squared_error(yte, y_pred))
    r2 = r2_score(yte, y_pred)
    print(f"Regression performance: MAE={mae:.2f}, RMSE={rmse:.2f}, R2={r2:.3f}")

    # Classification quick
    clf_pipe = Pipeline([('pre', preproc), ('rf', RandomForestClassifier(n_estimators=80, max_depth=10, n_jobs=-1, random_state=RSEED, class_weight='balanced'))])
    Xtrc, Xtec, ytrc, ytec = train_test_split(X, y_clf, test_size=0.2, random_state=RSEED, stratify=y_clf)
    print("Training classification model (fast RF) ...")
    clf_pipe.fit(Xtrc, ytrc)
    ycpred = clf_pipe.predict(Xtec)
    ycpred_prob = clf_pipe.predict_proba(Xtec)[:,1]
    acc = accuracy_score(ytec, ycpred); f1 = f1_score(ytec, ycpred)
    roc = roc_auc_score(ytec, ycpred_prob)
    print(f"Classification performance: Acc={acc:.3f}, F1={f1:.3f}, ROC-AUC={roc:.3f}")

    # Save model performance summary
    perf = pd.DataFrame({
        "task":["regression","classification"],
        "MAE":[mae, None],
        "RMSE":[rmse, None],
        "R2":[r2, None],
        "Accuracy":[None, acc],
        "F1":[None, f1],
        "ROC-AUC":[None, roc]
    })
    perf.to_csv("outputs/model_performance_summary.csv", index=False)
    display(perf)
else:
    print("Not enough columns for quick modeling or 'price' missing.")

# 6. Save a short final insights file (auto)
insights = []
insights.append("Q1: Room types distribution saved (outputs/q1_room_type_counts.png).")
if 'neighbourhood_group' in df.columns:
    insights.append("Q2: Neighbourhood group counts saved (outputs/q2_neighbourhood_group_listings.png).")
insights.append("Q3: Top neighbourhoods by average price saved (outputs/q3_top10_neighbourhoods_mean_price.png).")
if 'construction_year' in df.columns:
    insights.append("Q4: Construction year vs price plot saved (outputs/q4_construction_vs_price.png).")
if 'host_id' in df.columns:
    insights.append("Q5: Top hosts table saved to notebook view.")
insights.append("Q6: Host verification vs review rate plot saved (outputs/q6_host_verified_vs_reviews.png).")
insights.append("Q7: Price vs service fee scatter saved (outputs/q7_price_vs_service_fee.png).")
insights.append("Q8: Review-rate by neighbourhood & room type saved (outputs/q8_review_by_neighbourhood_group.png and q8_review_by_room_type.png).")
insights.append("Q9: Availability vs host listings plots saved (outputs/q9_*.png).")

with open("outputs/final_insights_questions.txt","w") as f:
    f.write("\n".join(insights))

print("\nAll question analyses complete. Files saved to 'outputs/' and cleaned CSV in 'dataset/'.")
print("Tip: open the Files pane in Colab to download any images/CSVs.")
```
